##Zertan datza komunitate detekzioa?
Komunitateen detekzioaren kontzeptua sareen zientzian sortu da, sistema konplexuen barruan taldeak aurkitzeko metodo gisa, grafiko batean irudikatuta. Deskonposizio-metodo tradizionalenek bloke diagonalen edo triangeluarren egitura zorrotza bilatzen dute, baina komunitateen detekzio-metodoek talde bereko nodoen lotura esanguratsua(handia) eta talde desberdinetako nodoen arteko lotura txikia duten azpisareak aurkitzen dituzte. Beraz, Community Detection problema sendoki elkar erlazionaturik dauden ertzen taldeak bilatzean datza talde desberdinetako ertzen loturak sendotasun txikia dutelarik. Erlazio hori ertzek dituzten propietateen bidez ezartzen da.

##Zertarako erabili daiteke komunitate detekzioa?

Gauza askotarako erabili daiteke, hala nola,  komunitateak detektatzeko teknikak baliagarriak dira sare sozialetako algoritmoek interes komunak dituzten pertsonak aurkitzeko eta estuki konektatuta edukitzeko. Beste adibide bat honako izan daiteke: komunitateen detekzioa ikaskuntza automatikoan erabiltzea antzeko propietateak dituzten taldeak detektatzeko eta hainbat arrazoirengatik taldeak ateratzeko.

##Komunitate bat definitzen

Komunitate kontzeptua hainbat alditan aipatu dugu, baina, nola defini genezake komunitate bat?

Esan genezake komunitate bat sare askoren propietate bat dela, zeinak horietako sare zehatz batek komunitate ugari izan ditzake eta komunitate baten barruko nodoak dentsitate handiz konektatuta egoten diren.

##Clustering vs Komunitate detekzioa

Komunitateen detekzioa clustering-aren antzekoa dela esan daiteke. Clusteringa ikaskuntza automatikoko teknika bat da, eta antzeko datu-puntuak kluster berean biltzen dira, bere atributuetan oinarrituta. Clusteringa sareei aplika dakiekeen arren, eremu zabalagoa da gainbegiratu gabeko ikaskuntza automatikoan, hainbat atributu motaz arduratzen dena. Bestalde, komunitateen detekzioa sareak aztertzeko bereziki diseinatuta dago, eta hori ertz izeneko atributu mota bakar baten mende dago. Gainera, clustering algoritmoek nodo periferiko bakarrak eta kide izan beharko luketen komunitateak bereizteko joera dute. Hala ere, clustering-teknikak eta komunitateak detektatzeko teknikak sareak aztertzeko arazo askori aplika dakizkieke, eta hainbat alde on eta txar plantea ditzakete eremuaren arabera.

##Komunitate detekziorako teknikak

Komunitateak detektatzeko metodoak, oro har, bi motatan sailka daitezke: metodo aglomeratiboak eta metodo zatitzaileak. Metodo aglomeratiboetan, ertzak banan-banan gehitzen zaizkio nodoak baino ez dituen grafiko bati. Ertzak indartsuenetik ahulenera gehitzen dira. Metodo zatitzaileak metodo aglomeratiboen aurkakoak dira. Horietan, ertzak banan-banan ezabatzen dira grafiko oso batetik.

Sare jakin batean edozein komunitate egon daiteke, eta komunitate horiek tamaina desberdinekoak izan daitezke. Ezaugarri horiek direla eta, komunitateak detektatzeko prozedura oso zaila da. Guk gure helburu-funtzio propioa proposatzen dugu. Honen ideia komunitate bakoitzeko elkarrekin konektatuta dauden nodoen ertzen batura egitea da. Hau formula matematiko baten bidez horrela adieraziko genuke:

$$f(x)=\sum_{i=1}^n\sum_{j=i}^n 
\begin{cases} 
      w_{ij} & x_{i} = x_{j} \\
      0 & x_{i} \neq x_{j}
\end{cases}
$$

Honen arazo nagusia da denak komunitate berdin batean sartzeko joera duela, eta honen bidez helburu funtzioko oso balio altuak lortzen ditu. Beraz, honekin hurrengo bi puntuetan azaltzen dena lortzen dugu:
- Komunitate oso handi bat probabilitate askorekin oso kohesio txikia izateko.

Joera honen ondorioz, helburu funtzio hau baztertzea erabaki dugu. Horren ordez, problema berdinean aplikatu diren modularitate funtzio bat aukeratzea erabaki dugu.


Community Detection problemari buruzko informazioa bilatu ondoren, https://arxiv.org/abs/cond-mat/0408187 link-ean dagoen paperrean agertzen den helburu funtzioa erabiltzea erabaki dugu, hurrengo ekuazioak deskribatzen duena:

$$f(x)= \frac{1}{2m}\sum_{ij} \left[  A_{ij} - \frac{ k_ik_j } {2m}  \right] \delta(c_i,c_j)
$$

Helburu funtzio honen bidez komunitate baten trinkotasuna neurtu dezakegu, izan ere, komunitate bateko barruko ertzen batura handia izatea nahi dugu, beste komunitateekin dituen ertzen batura ahalik eta txikien mantenduz. Helburu funtztio honek hau egitea ahalbidetzen digu.

$f(x)$ modularitatea da, $\delta(c_i,c_j)$ funtzioak i nodoa j nodoaren auzokidea den esaten digu. 0 itzuliko du auzokidea ez bada bestela 1ekoa itzuliko du. $A_{ij}$ i nodotik j nodora dagoen ertzaren pisua adierazten digu. $k_i$ da i nodoa beste nodoekin konektatuta dagoen ertzen batura. $m$ grafoko ertz guztien batura adierazten digu, $m = \frac{1}{2} \sum_{ij}A_{ij}$ izanik.

Modularitatea sare bat zatitzea erabakitzen den moduaren propietate bat da: zatituta ez dauden sareek eta nodo bakoitza beren komunitatean jartzen dutenek zero modularitatea izango dute.

Gainera, irakurritako paper-ak honakoa esaten digu, erabilitako network-a ausazkoa baldin bada orduan modularitate funtzioak eman beharreko emaitza 0 izan beharko luke. Beraz, zero baino txikiagoak diren balioak emaitza txarrak izango dira. Zero baino handiagoak diren balioak ausazkotasunetik desbideratzeak (onera)adierazten dituzte, eta, praktikan, 0,3 inguruko balioa sare batean komunitate esanguratsu bat dagoela esan nahi du.

Hau da helburu funtzioa inplementatuta:

```python
def modularitateaOld(G, soluzioa, weight='weight'):
    value = 0
    nodoak = list(G.nodes)
    m = G.size(weight="weight")/2
    for i in range(len(soluzioa)):
        for j in range(len(soluzioa)):
            if soluzioa[i] != soluzioa[j]:
                continue
                
            lag = G[nodoak[i]]
            if  nodoak[j] not in lag: #Ez dira auzokideak
                pisua = 0
            else:
                pisua = lag[nodoak[j]][weight]  
            
            degree_i = G.degree(nodoak[i]) 
            degree_j = G.degree(nodoak[j]) 
            value += pisua - degree_i*degree_j/(2*m)
    return value/(2*m)
```

Helburu funtzio hau oso motela da, eta horregatik bektorizatzea erabaki genuen. Ideia da pisuak eta $\frac{k_ik_j}{2m}$ balioak matrizeetan sartu eta np bidez hauen kenketa egitea. Modularitate funtzio bektorizatua UA taldearen eta Txintxarri eta kunplitu taldearen kolaborazioa izan da. Matrizeen hasieraketak horrela egiten ditugu: 

```python
if __name__ == "__main__":
  #Nodoen izenen lista (indizetik izenera pasatzeko)
  nodoak = list(G.nodes)

  #Grafoko ertz guztien pisuen batura
  m = G.size(weight="weight") 

  #Nodoen izenetatik indizera pasatzeko hiztegia
  indizeak = dict()
  ind = 0
  for nodo in nodoak:
    indizeak[nodo] = ind
    ind+=1

  #Lortu nodo batetik besterako ertzen pisuen matrizea (ertza ez badago pisua=0)
  pisuak = np.zeros((n,n))
  for i in range(n):
      for j in range(n):
        lag = G[nodoak[i]]
        if  nodoak[j] in lag: #dira auzokideak
            pisuak[i,j] = lag[nodoak[j]]["weight"]

  #Nodoen graduak eskuratzeko. Dena den, guk graduak erabili ordez nodoari konektatutako ertzen pisuen baturak erabiliko ditugu modularitatean.
  # graduak = []
  # for i in range(n):
  #   graduak.append(G.degree(nodoak[i]))

  #Nodo baten gradua beregana konektatutako ertz guztien pisuen batura kontsideratuko dugu
  graduak = []
  for i in range(n):
    graduak.append(np.sum(pisuak[i]))

  #Modularitate funtziorako aurrekalkulatu i eta j nodo pare posible guztien graduen biderketak 
  graduMul = np.zeros((n,n))
  for i in range(n):
      for j in range(n):
        graduMul[i,j] = graduak[i]*graduak[j]/(2*m)
```

Eta horrela geratuko litzateke gure modularitate funtzioa bektorizatuta:

```python
import numpy as np
def modularitatea(G, solution):
    n = G.number_of_nodes()
    ii, jj = np.ogrid[:n, :n]
    value = np.sum(np.where(solution[ii]==solution[jj], pisuak[ii,jj]-graduMul[ii,jj], 0)) 
    return value/(2*m)
```
